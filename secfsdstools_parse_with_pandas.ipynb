```python
num = pd.read_csv(given_path+'num.txt', sep='\t')
num
```

```python
adsh_tag_ddate_to_value = {}
total_count = 0
errored_idx = []

for idx, (adsh, tag, version, coreg, qtrs, ddate, uom, value) in enumerate(
    zip(num.adsh, num.tag, num.version, num.coreg, num.qtrs, num.ddate, num.uom, num.value)):
    if (adsh, tag, version, coreg, qtrs, ddate, uom) in adsh_tag_ddate_to_value:
        cur_value = adsh_tag_ddate_to_value[adsh, tag, version, coreg, qtrs, ddate, uom]
        if cur_value != value:
            print('Duplicate exists, but value is different!')
            total_count += 1
            print(total_count)
            errored_idx += [idx]
    else:
        adsh_tag_ddate_to_value[adsh, tag, version, coreg, qtrs, ddate, uom] = value
```

```python
sub = pd.read_csv(given_path+'sub.txt', sep='\t')
sub
```

```python
import numpy as np
```

```python
sub['period'] = sub.period.map(lambda x: str(x)[:-2] if x != np.nan else x)
sub['filed'] = sub['filed'].map(lambda x: str(x))
```

```python
adsh_to_cik = {}
for adsh, cik in zip(sub.adsh, sub.cik):
    adsh_to_cik[adsh] = cik
```

```python
cik_to_adsh = {}
for adsh, cik in adsh_to_cik.items():
    cik_to_adsh[cik] = adsh
```

```python
cik_to_accepted_date = {}
for cik, accepted_date in zip(sub.cik, sub.accepted):
    cik_to_accepted_date[cik]=accepted_date
    
cik_to_period_ending = {}
for cik, period_ending_date in zip(sub.cik, sub.period):
    cik_to_period_ending[cik]=period_ending_date
```

```python
pre['cik'] = pre.adsh.map(lambda x: adsh_to_cik[x])
pre['accepted_date'] = pre.cik.map(lambda x: cik_to_accepted_date[x])
pre['period'] = pre.cik.map(lambda x: cik_to_period_ending[x])
```
